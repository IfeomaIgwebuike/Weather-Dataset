```python
import panel as pn
import pandas as pd
import hvplot.pandas

pn.extension()
```

```python
df = pd.read_csv('autompg.csv')
df.head()
```

```python
df.hvplot.line(y = ['mpg', 'accel'],
              legend = 'left',
              xlim = [-20, 400],
              xlabel = 'data').opts(xticks = [50, 200, 250])
```

# WEATHER LINE CHAT

```python
sf = pd.read_csv('WeatherAssign.csv')
sf.head()
```

```python
sf.hvplot.line(y = ['Precipitation', 'Temperature'],
              legend = 'left',
              xlim=(sf['Precipitation'].min(),sf['Precipitation'].max()),
              xlabel = 'data').opts(xticks = [50, 200, 250])
```

# BAR CHART

```python
df.loc[0:100, 'mpg'].hvplot.bar(height = 400,
                               legend = 'top_left',
                               rot = 60,
                               ylabel = 'Y',
                               bar_width = 0.8,
                               title = 'X vs Y')
```

```python
# agg. or sum.
table = df.groupby('mfr')[['mpg','cyl', 'displ', 'accel']].mean()
table
```

```python
#table.hvplot.bar(y = 'mpg', rot = 90)
table.hvplot.bar(y = ['mpg', 'accel'], rot = 90)
```

```python
table2 = df.groupby(['yr', 'origin'])[['cyl']].mean()
table2.head()
```

```python
table2.hvplot.bar(legend = 'top_left',
                  rot = 90,
                 bar_width = 0.5,
                 fontscale = 1.5,
                 stacked = True,
                 #color = ['black', 'silver', 'green'],
                 colormap = 'dark2')
```

```python
# BARCHART FOR WEATHER DATA
```

```python
sf = pd.read_csv('WeatherAssign.csv')
sf.head()
```

```python
sf.loc[0:100, 'Precipitation'].hvplot.bar(height = 400,
                               legend = 'top_left',
                               rot = 60,
                               ylabel = 'Y',
                               bar_width = 0.8,
                               title = 'X vs Y')
```

```python
# agg. or sum.

table = sf.groupby('Year')[['Precipitation','Min_Temperature', 'Max_Temperature', 'Temperature']].mean() 
table
```

```python
#table.hvplot.bar(y = 'mpg', rot = 90)
table.hvplot.bar(y = ['Precipitation', 'Temperature'], rot = 90)
```

```python
table2 = sf.groupby(['Year', 'Country'])[['Temperature']].mean()
table2.head()
```

```python
table2.hvplot.bar(legend = 'top_left',
                  rot = 90,
                 bar_width = 0.5,
                 fontscale = 1.5,
                 stacked = True,
                 #color = ['black', 'silver', 'green'],
                 colormap = 'dark2')
```
